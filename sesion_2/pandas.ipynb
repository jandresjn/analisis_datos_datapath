```python
import pandas as pd

# Crear un DataFrame más complejo con varias columnas
data = {
    'grupo': ['A', 'B', 'A', 'A', 'B', 'C', 'C', 'A'],
    'subgrupo': ['X', 'Y', 'X', 'Z', 'Y', 'Z', 'X', 'Z'],
    'valor': [10, 25, 30, 40, 5, 15, 50, 20],
    'cantidad': [1, 2, 3, 1, 2, 1, 5, 2]
}

df = pd.DataFrame(data)

# Agrupamos por 'grupo' y calculamos el promedio de las columnas 'valor' y 'cantidad'
promedio = df.groupby('grupo').agg({
    'valor': 'mean',  # Promedio de la columna 'valor'
    'cantidad': 'sum'  # Suma de la columna 'cantidad'
})

# Mostramos el resultado
promedio
```
